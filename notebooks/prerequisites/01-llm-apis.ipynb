### Load Environment

```python
from dotenv import load_dotenv
import os

load_dotenv("../../.env")
```

### OpenAI Client

```python
import openai

response = openai.chat.completions.create(
    model="gpt-5-nano",
    messages=[
        {"role": "system", "content": "You are a helpful assistant."},
        {"role": "user", "content": "can you print 10 random words?"}
    ]
)

print(response.choices[0].message.content)
```

```python
response.usage
```

```python
response = openai.chat.completions.create(
    model="gpt-5-nano",
    messages=[
        {"role": "system", "content": "You are a helpful assistant."},
        {"role": "user", "content": "can you print 10 random words?"}
    ],
    reasoning_effort="minimal" # can also set to low, etc...
)

print(response.choices[0].message.content)
```

### Google genai Client

```python
from google import genai

gemini_client = genai.Client(api_key=os.getenv("GEMINI_API_KEY"))

response = gemini_client.models.generate_content(
    model="gemini-2.5-flash",
    contents = ["You are a helpful assistant",
                "can yoiu print 15 random words?"]
)

print(response.text)
```

### Groq Client

```python
from groq import Groq

client = Groq(api_key=os.getenv("GROQ_API_KEY"))

chat_completion = client.chat.completions.create(
    messages = [
        {"role": "system", "content": "You are a helpful assistant."},
        {"role": "user", "content": "can you print 20 random words?"}
    ],
    model="llama-3.3-70b-versatile"
)

print(chat_completion.choices[0].message.content)
```
